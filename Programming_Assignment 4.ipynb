1. Write a Python Program to Find the Factorial of a Number?

```python
num = int(input("Enter a number: "))
factorial = 1
for i in range(num, 1, -1):
    factorial *= i
print(f'Factorial of {num} is {factorial}')
```

2. Write a Python Program to Display the multiplication Table?

```python
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f'{num} x {i} = {num * i}')
```

3. Write a Python Program to Print the Fibonacci sequence?

```python
num = int(input("Enter a number: "))
fibonacci = [0, 1]
for i in range(2, (num + 1)):
    fibonacci.append((fibonacci[i - 1] + fibonacci[i - 2]))
    
print(fibonacci)
```

4. Write a Python Program to Check Armstrong Number?

```python
num = input("Enter a number: ")
sum_of_digits_cubed = sum([(int(i) ** 3) for i in num])
if int(num) == sum_of_digits_cubed:
    print(f'{num} is Armstrong number')
else:
    print(f'{num} is not Armstrong number')
```

5. Write a Python Program to Find Armstrong Number in an Interval?

```python
start = int(input('Enter the starting number: '))
end = int(input('Enter the ending number: '))
armstrong_nums = []
for num in range(start, (end + 1)):
    sum_of_digits_cubed = sum([(int(i) ** 3) for i in str(num)])
    if num == sum_of_digits_cubed:
        armstrong_nums.append(num)

print(armstrong_nums)
```

6. Write a Python Program to Find the Sum of Natural Numbers?

```python
n = int(input("Enter a number: "))
sum_of_nat_nums = (n * (n + 1)) / 2
print(f'sum of first {n} natural numbers is {sum_of_nat_nums}')
```
